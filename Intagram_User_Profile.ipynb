```python
import pandas as pd
import json
from os import listdir
from os.path import isfile, isdir
```

```python
import warnings
warnings.filterwarnings('ignore')
```

```python
def ls1(path):    
    return [obj for obj in listdir(path) if isfile(path + obj)]
```

```python
data =[]
files=ls1("C:/Users/Hammer/Jupyter/profiles/")
for file in files:
    data.append(file)
```

```python
rutas =[]
for i in data:
    ruta = 'profiles/'+i
    rutas.append(ruta)
```

```python
datw=[]
for i in rutas:
    with open(i, 'rb') as f:
        da = json.load(f)
        datw.append(da)
```

```python
df = pd.DataFrame.from_dict(datw, orient='columns')
```
